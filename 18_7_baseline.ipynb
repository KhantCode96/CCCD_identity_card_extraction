```python
from pathlib import Path
import subprocess
import json
import random
import pandas as pd

SCRIPT = Path(
    "/content/drive/MyDrive/YOLOV8-VIET_OCR/extract_cccd.py"
)

IMAGE_DIR = Path(
    "/content/drive/MyDrive/YOLOV8-VIET_OCR/"
    "CCCD_dataset-8/test/images"
)

YOLO_WEIGHTS = Path(
    "/content/drive/MyDrive/YOLOV8-VIET_OCR/"
    "training_output/fresh_run/cccd_yolov8_best.pt"
)

OUTPUT_DIR = Path(
    "/content/drive/MyDrive/YOLOV8-VIET_OCR/"
    "training_output/fresh_run/batch_test_15"
)

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

extensions = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}

all_images = sorted(
    image for image in IMAGE_DIR.iterdir()
    if image.suffix.lower() in extensions
)

assert SCRIPT.is_file(), f"Script not found: {SCRIPT}"
assert YOLO_WEIGHTS.is_file(), f"Weights not found: {YOLO_WEIGHTS}"
assert len(all_images) >= 15, f"Only {len(all_images)} test images found"

random.seed(42)
selected_images = random.sample(all_images, 15)

records = []

for index, image_path in enumerate(selected_images, start=1):
    debug_path = OUTPUT_DIR / f"{image_path.stem}_debug.jpg"
    json_path = OUTPUT_DIR / f"{image_path.stem}_result.json"

    print(f"[{index}/15] {image_path.name}")

    command = [
        "python",
        str(SCRIPT),
        "--image",
        str(image_path),
        "--weights",
        str(YOLO_WEIGHTS),
        "--save-debug",
        str(debug_path),
    ]

    process = subprocess.run(
        command,
        capture_output=True,
        text=True,
    )

    if process.returncode != 0:
        print("  FAILED")
        print(process.stderr)

        records.append({
            "image": image_path.name,
            "status": "failed",
            "id": "",
            "birth": "",
            "name": "",
            "warnings": process.stderr[-500:],
        })
        continue

    # The script may print warnings before the JSON,
    # so locate the beginning of the final JSON object.
    json_start = process.stdout.find("{")

    try:
        result = json.loads(process.stdout[json_start:])

        json_path.write_text(
            json.dumps(result, ensure_ascii=False, indent=2),
            encoding="utf-8",
        )

        print(
            f"  SUCCESS | ID={result['id']} | "
            f"Birth={result['birth']} | Name={result['name']}"
        )

        records.append({
            "image": image_path.name,
            "status": "success",
            "id": result["id"],
            "birth": result["birth"],
            "name": result["name"],
            "warnings": "; ".join(result["warnings"]),
            "detected_classes": len(result["detections"]),
            "debug_image": str(debug_path),
        })

    except Exception as error:
        print(f"  JSON PARSE FAILED: {error}")

        records.append({
            "image": image_path.name,
            "status": "json_parse_failed",
            "id": "",
            "birth": "",
            "name": "",
            "warnings": str(error),
        })

report = pd.DataFrame(records)

csv_path = OUTPUT_DIR / "batch_results.csv"
report.to_csv(csv_path, index=False, encoding="utf-8-sig")

print("\nBatch test complete.")
print("Results folder:", OUTPUT_DIR)
print("CSV report:", csv_path)

display(report)
```
